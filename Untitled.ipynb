```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
# plot a scatter plot between total_bill and tips
```

```python
sns.relplot(x='total_bill',y='tip' ,data=data , kind='scatter')
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

```python
# plotting on categorical data
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=data,jitter=0)
```

```python
sns.catplot(x='day',y='tip',kind='swarm' , data = data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm' , data = data , hue='sex')
```

```python
sns.catplot(x='day',y='tip' , data = data )
```

```python
# boxplot
sns.boxplot(data['total_bill'])
```

```python
sns.catplot(x='day' , y='total_bill' , hue='sex' , kind='box' , data = data)
```

```python
# violin plot
# to compare  the distribution
```

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin', hue = 'sex',split=True,data=data)
```

```python
# barplot and count plot to get central tendency of the data
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',estimator=np.median , data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',estimator=np.var , data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',estimator=np.std , data=data)
```

```python
sns.catplot(x='smoker', hue='smoker',kind='count' , data=data)
```

```python
# heatmap : indivisual values represented in matrix with different colours
```

```python
flights = pd.read_csv('flights.csv')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x , cbar=False) #cbar false is false to remove the scale
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5) # give the space between the boxes
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 ,annot=True ,fmt='d')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 ,annot=True ,fmt='d' , cmap='viridis')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 ,annot=True ,fmt='d' , cmap='summer')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 ,annot=True ,fmt='d' , cmap='winter')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 ,annot=True ,fmt='d' , cmap='BuPu')
```

```python
sns.clustermap(x)
# make the cluster of similar columns and rows
```

```python
sns.clustermap(x , metric='correlation')
```

```python
# joint point : for multivariate analysis
# Scatter plot
# Hexbin Plots
# KdePlots
# RegPlot
```

```python
sns.jointplot(x='total_bill',y='tip' , data = data )
```

```python
sns.jointplot(x='total_bill',y='tip' ,kind='hex', data = data )
```

```python
sns.jointplot(x='total_bill',y='tip' ,kind='kde', data = data )
```

```python
sns.jointplot(x='total_bill',y='tip' ,kind='reg', data = data )
```

```python
sns.jointplot(x='total_bill',y='tip' ,kind='resid', data = data )
```
